** Whisker 3 means increasing length of whisker line so that more of data is considered as normal  **

````python
def count_outlier(s,whis=1.5):
    q1,q3=np.percentile(s,[25,75])
    iqr=q3-q1
    upper_bound=q3+(whis*iqr)
    lower_bound=q1-(whis*iqr)
    count=len(s.loc[(s<lower_bound)|(s>upper_bound)])
    return count

print(count_outlier(df['Sales Price']))
print(count_outlier(df['Sales Price'],whis=3))
````

````python
def remove_outliers(df, col):
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3-q1
  lower_bound = q1 - 3 * iqr
  upper_bound = q3 + 3 * iqr

  outlier_mask = (df[col] < lower_bound) | (df[col] > upper_bound)
  return df[~outlier_mask]

remove_outliers(df, 'Sales Price').shape
````

````python
print(remove_outliers(df,'Profit').shape)
````

````python
def remove_outlier(df, col, whis=1.5):
    q1, q3 = np.percentile(df[col], [25, 75])
    iqr = q3 - q1
    low, up = q1 - (iqr*whis), q3 + (iqr*whis)
    data_new = df.loc[(df[col] >= low) & (df[col] <= up)]
    return data_new

print(remove_outlier(df, 'Sales Price').shape)
print(remove_outlier(df, 'Sales Price', whis=3).shape)
````

# Secton 2.2  HYPOTHESIS TESTING


*Business QUES : How do discounts affect profit*

** TO ANS THIS QUES perfect tool is OLS(ORDINARY LEAST SQUARE) REGRESSION MODEL  **

````python
import statsmodels.api as sm
# Defining the independent variable x and dependent variable y
model_data=df[['Discount','Total Profit']].dropna()
x=model_data['Discount']
y=model_data['Total Profit']

#add a constant to the independent variable for the intercept term as it is not done automatically in this library
X = sm.add_constant(x)

#create fit OLS MODEL
model = sm.OLS(y, X).fit()

# Print the summary of the model
print(model.summary())
````

**H0(Null Hypothesis)= Discount has no real effect on Profit. Negative Trend we see is random or by chance**


**HA(Alternate Hypethesis)=Offering a Discount do have a negative effect on Profit**


**We get P-VALUE=0.000** .It is far less than 5% .

 We reject NULL HYPOTHESIS that discount have no effect.

We conclude that our ALTERNATIVE HYPOTHESIS is correct that **discount do affect Profit.**



Then QUES arises

**How Big is the Impact? (The Coefficients and Confidence Interval) **

*Discount Co-efficient=  -21.9
 which means for every 1 unit of discount we loss 22 in profit .

BUSS QUES-Is the business driven by new customer or keep existing ones?

The core of this analysis to give every single order a label 'New' or 'Existing' ones.
For this we need to find date of theeir ver first purchase.


1)Find the First Order: For every customer, we find the earliest Order Date in their history.
Compare Every Purchase: We then look at every single transaction they've ever made.

2)Apply the Label:
If a transaction's month is the same as their first-ever purchase month, that sale is from a 'New' customer.
If a transaction's month is later than their first-ever purchase month, that sale is from an 'Existing' customer.

**What’s the main logic?**

*We are doing a cohort-like analysis:*

Identify when customers were acquired.

Label each order as New customer purchase or Existing customer purchase.

Aggregate monthly revenue split by New vs Existing customers.**






````python
df['acquistion date']=df.groupby('Customer ID')['Order Date'].transform('min')
df['acquistion month']=df['acquistion date'].dt.to_period('M')
df['Order Month']=df['Order Date'].dt.to_period('M')
df['order year']=df['Order Date'].dt.year
````

````python
df['customer_type']=np.where(df['Order Month']==df['acquistion month'], 'New','Existing')
````

````python
min_date=df['Order Date'].min()
max_date=df['Order Date'].max()
start_year=min_date.year
end_year=max_date.year
````

````python
min_date=df['Order Date'].min()
max_date=df['Order Date'].max()
start_year=min_date.year
end_year=max_date.year

filtered_data_for_plot = df[(df['Order Date'].dt.year >= start_year) & (df['Order Date'].dt.year <= end_year)].copy()

revenue_by_type_filtered = filtered_data_for_plot.groupby(['Order Month', 'customer_type'])['Total Sales'].sum().unstack().fillna(0)

revenue_by_type_filtered.head()
````

# The Strategic action plan
How can we visualize our entire product strategy at single chart

Q.1]Which product is more in demand ?

Q.2]Which products are sold steadily without much fuss?

Q.3]Which products are new , but might find potential i promote them?

Q.4]Which products are taking space , must be discontinued?

This code creates a strategic map—often called a **BCG Matrix**—to solve this exact dilemma for our Superstore business. It plots every single product on a grid to instantly reveal its strategic role.

````python
product_summary=df.groupby('Product Name').agg(Total_Quantity_sold=('Quantity','sum'),profit_per_unit=('Total Profit','mean'),catagory=('Category','first')).dropna()
product_summary.head(10)
````

````python
g = sns.FacetGrid(data=product_summary, col='catagory', hue='catagory', height=6)
g.map(sns.scatterplot, 'Total_Quantity_sold', 'profit_per_unit')

for ax in g.axes.flat:
    vol_median=product_summary['Total_Quantity_sold'].median()
    effi_median=product_summary['profit_per_unit'].median()
    ax.axvline(vol_median,color='red',linestyle='--')
    ax.axhline(effi_median,color='red',linestyle='--')
    ax.set_xlabel('Total Quantity Sold')
    ax.set_ylabel('Profit Per Unit')

g.add_legend()
plt.show()
````
